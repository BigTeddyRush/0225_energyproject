```python
import requests
import yaml
```

```python
def load_config(config_file="config.yaml"):
    """Load configuration from a YAML file."""
    with open(config_file, "r") as file:
        config = yaml.safe_load(file)
    return config
```

```python
config = load_config()
filter_ids = config.get("FILTER_IDS", [])
regions = config.get("REGIONS", [])
resolutions = config.get("RESOLUTIONS", [])    


for filter_id in filter_ids:
    for region in regions:
        for resolution in resolutions:
            timestamp_url = f"https://www.smard.de/app/chart_data/{filter_id}/{region}/index_{resolution}.json"
            print(f"Requesting timestamps from: {timestamp_url}")
            response = requests.get(timestamp_url, timeout=60)
            timestamps = response.json()["timestamps"]

            # Check if timestamps were returned
            if timestamps:
                # Pick the latest timestamp (or any you want)
                print(timestamps)
                latest_timestamp = timestamps[-1]
                print(f"Latest timestamp: {latest_timestamp}")

                # Construct URL for time series data
                timeseries_url = (
                    f"https://www.smard.de/app/chart_data/{filter_id}/{region}/"
                    f"{filter_id}_{region}_{resolution}_{latest_timestamp}.json"
                )
                print(f"Requesting time series data from: {timeseries_url}")
                ts_response = requests.get(timeseries_url, timeout=60)
                raw_timeseries_data = ts_response.json()["series"]
                print("Time series data sample:", raw_timeseries_data)
            else:
                print("No timestamps found.")
```
